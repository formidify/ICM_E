```python
#first import pandas to fetch data
import pandas as pd
def fetch_data(route):
    data = pd.read_csv(route)
    return data
```

```python
fsi = fetch_data("fsi-2015.csv");
elavation = fetch_data("elevation.csv");
population = fetch_data("population_displacement.csv");
arable_land = fetch_data("percent_arable_land.csv");
undernourishment = fetch_data("undernourishment.csv");
corruption = fetch_data("control_of_corruption.csv");
gov_eff = fetch_data("government_effectiveness.csv");
political_stability = fetch_data("political_stability.csv");
precipitation = fetch_data("precipitation.csv");
rule_of_law =  fetch_data("rule_of_law.csv");
temperature = fetch_data("temperature.csv");
voice = fetch_data("voice_and_accountability.csv");
```

```python
fsi = fsi.drop(["Year"],axis=1)
fsi.describe()
```

```python
fsi.head(10)
```

```python
h = fetch_data("Hunger.csv")
```

```python
h.head(3)
```

```python
fsi["C"] = fsi["C1: Security Apparatus"]+fsi["C2: Factionalized Elites"]+fsi["C3: Group Grievance"]
fsi
```

```python
fsi["P"] = fsi["P1: State Legitimacy"]+fsi["P2: Public Services"]+fsi["P3: Human Rights"]
fsi.head(108)
```

```python
fsi["E"]=fsi["E1: Economy"]+fsi["E2: Economic Inequality"]+fsi["E3: Human Flight and Brain Drain"]
```

```python
fsi["S"]=fsi["S1: Demographic Pressures"]+fsi["S2: Refugees and IDPs"]
```

```python
fsi["X"] = fsi["X1: External Intervention"]
```

```python
fsi.describe()
```

```python
fsi = fsi.dropna(subset=['S'])
```

```python
data_cat =  fsi[["Country","C", "E", "P", "S", "X"]]
```

```python
data_cat=data_cat.dropna(subset=['Country'])
```

```python
data_cat
```

```python
merged = pd.merge(h,data_cat, on='Country', how='outer');
merged
```

```python
import pycountry

input_countries = merged["Country"]

countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_3

merged["code"] = [countries.get(country, 'Unknown code') for country in input_countries]
```

```python
merged = merged.merge(elavation, on='code',how='outer')

merged
```

```python
merged = merged.merge(population, on='code',how='outer')

merged
```

```python
merged = merged.merge(arable_land, on='code',how='outer')
merged = merged.merge(undernourishment , on='Country',how='outer')
merged = merged.merge(corruption , on='code',how='outer')
merged = merged.merge(gov_eff , on='code',how='outer')
merged = merged.merge(political_stability , on='code',how='outer')
merged = merged.merge(rule_of_law , on='code',how='outer')
merged = merged.merge(temperature , on='code',how='outer')
merged = merged.merge(voice , on='code',how='outer')
```

```python
merged
```

```python
merged.to_csv('merged.csv')
```
